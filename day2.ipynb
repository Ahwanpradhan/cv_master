```python
import cv2
import numpy as np
import argparse
import imutils
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('/content/post-10.jpg')
cv2_imshow(image)
```

```python
#get center of image
(h,w) = image.shape[:2]
(cx, cy) = (w//2, h//2)
```

```python
#rotate image by 45 degrees and scale 1
M = cv2.getRotationMatrix2D((cx,cy), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
cv2_imshow(rotated)
```

```python
rotated = imutils.rotate(image, 80)
cv2_imshow( rotated)
```
