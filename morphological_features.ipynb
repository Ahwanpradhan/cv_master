```python
#Erosion
import cv2
image=cv2.imread('/content/logo-social.png')
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(gray)
```

```python
for i in range(0, 10):
	eroded = cv2.erode(gray.copy(), None, iterations=i + 1)
	cv2_imshow(eroded)
	cv2.waitKey(0)
```

```python
#Dilation

import cv2
image=cv2.imread('/content/logo-social.png')
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
for i in range(0, 10):
	dilated = cv2.dilate(gray.copy(), None, iterations=i + 1)
	cv2_imshow(dilated)
	cv2.waitKey(0)
```
